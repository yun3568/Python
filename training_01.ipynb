```python
import pandas as pd

raw_data = {
    "date":["2023-01-01","2023-01-02","2023-01-03"],
    "price":[10.5,11.2,10.8],
    "vol":[1000,2000,1500]
}

df = pd.DataFrame(raw_data)
print(df)
```

```python
close_prices = df["price"]
print(close_prices)
print(type(close_prices))
```

```python
my_list = [10,20,30]
print("列表乘以2的结果：")
print(my_list * 2)

print("Series 乘以2的结果：")
print(close_prices * 2)
```

```python
df['amount'] = df['price'] * df['vol']
print(df['amount'])
```

```python
mask = df['vol'] > 1200
print(mask)
```

```python
high_vol_days = df[mask]
print(high_vol_days)
```

```python
df['ma2'] = df['price'].rolling(window=2).mean()
df.dropna(inplace=True)
print(df)
```
