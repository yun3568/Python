```python
#11.25
#essential operation
import pandas as pd
import numpy as np
obj = pd.Series([4.5, 7.2, -5.3, 3.6], index=['d', 'b', 'a', 'c'])
#重新排座次 reindex
obj2 = obj.reindex(['a','b','c', 'd', 'e'])
print("重新索引后：\n",obj2)
```

```python
#dropping entries
data = pd. DataFrame(np.arange(16).reshape((4,4)), #新
                   index=['Ohio', 'Colorado', 'Utah', 'New York'],
                   columns=['one', 'two', 'three', 'four'])
print('原始数据：',data)
#删除行
print("删除'Ohio':\n",data.drop('Ohio'))  #默认丢行（axis=0）
#删除列
print('删除 two 列',data.drop('two',axis=1)) #丢列记得加axis=1
```

```python
#selection with loc& iloc
#1.loc(名字)
print('使用loc：\n',data.loc['Colorado',['two','three']]) #选co一行及two three两列 
#2.iloc选（位置）
print('使用iloc:\n',data.iloc[1,[1,2]]) #co是第二行（从0开始数），'two''three'是第1,2列
```
