```python
#按照首字母添加字典  难
#传统方法
words = ['apple', 'bat', 'bar', 'atom', 'book']
by_letter = {}
for word in words:
    letter = word[0]
    if letter in by_letter:
        by_letter[letter].append(word)
    else:
        by_letter[letter] = [word]
by_letter
```

```python
#setdefault()
by_letter = {}
for word in words:
    letter = word[0]
    by_letter.setdefault(letter,[]).append(word)
by_letter
```

```python
#defultdict
from collections import defaultdict
by_letter = defaultdict(list)
for word in words:
    by_letter[word[0]].append(word)
dict(by_letter)
```

```python
#Nested List Comprehensions 嵌套列表推导式 [表达式 for 大循环 in 大列表 for 小循环 in 大循环]
some_tuples = [(1, 2, 3), (4, 5, 6), (7, 8, 9)]
flattened = [x for tup in some_tuples for x in tup]
flattened
```

```python
#详细拆解
flattened = []
for tup in some_tuples:
    for x in tup:
        flattened.append(x)
```

```python
#function
def my_function(x,y,z=1.5):  #x,y是位置参数，z是关键字参数有默认值
    if z > 1:
        return z * (x+y)
    else:
        return z / (x + y)
result_a = my_function(5,6,z=0.7)
result_b = my_function(3.14,7,3.5)
print(f'结果A：{result_a}')
print(f'结果B：{result_b}')
```

```python
#Functions are Objects
import re  #正则表达式
states=['Alabama','Georgia!','Georgia','georgia','FlOrIda', 'south   carolina##', 'West virginia?']
def remove_punctuation(value):
    return re.sub('[!#?]','',value)  #re,sub(找什么，换成什么，在哪里找)
clean_ops = [str.strip#去除空格,remove_punctuation,str.title#首字母大写]
def clean_strings(strings,ops):
    result = []
    for value in strings:
        for function in ops:
            value = function(value)
        result.append(value)
    return result
clean_strings(states,clean_ops)
```

```python
#Lambda Functions 匿名函数
#用途：有时候一个函数只需要用一次，没必要起名字
#lambda 参数：返回值（不需要写return）
strings = ['foo', 'card', 'bar', 'aaaa', 'abab']
strings.sort(key=lambda x: len(set(x)))
strings
```

```python
#generator yield代替return 表示生成器  刚开始不会执行
def squares(n=10):
    print('Generating squares from 1 to {0}'.format(n ** 2))
    for i in range(1,n+1):
        yield i ** 2
gen = squares()
```

```python
#挤出第一个数
print(next(gen))
```

```python
#所有输出
for i in gen:
    print(i,end='')
```
