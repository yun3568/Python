```python
import pandas as pd
# 玩具数据
student_data = [
    {'name': '张三', 'math': 90, 'english': 85},
    {'name': '李四', 'math': 75, 'english': 92},
    {'name': '王五', 'math': 88, 'english': 80}
]
df_students = pd.DataFrame(student_data)
print(df_students)
print(type(df_students))
```

```python
print(df_students.columns)
math_scores = df_students['math']
print(math_scores)
print(type(math_scores))
```

```python
#实验A
df_students = pd.DataFrame(student_data)
math_scores_series = df_students['math']
math_scores_list = [90,75,88]
print("实验A(列表list):")
try:
    print(math_scores_list + 5)
except TypeError as e:
    print(f" ->报错了！错误信息：{e}")
    print(" ->结论：Python 列表不支持‘批量’数学运算。")

#实验B
print('\n实验 B (序列 Series)：')
print(math_scores_series + 5)
```

```python
english_scores_series = df_students['english']
df_students['total_score'] = math_scores_series + english_scores_series
print(df_students)
```

```python
print(df_students[df_students.math >= 80])
```
