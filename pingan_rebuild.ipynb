```python
import pandas as pd
import akshare

df = akshare.stock_zh_a_hist(symbol= '000001',period='daily',start_date='20230101',end_date='20231231',adjust='qfq')
print(df.head())
print(df.columns)
```

```python
df['MA5'] = df['收盘'].rolling(window=5).mean()
df['MA20'] = df['收盘'].rolling(window=20).mean()
Signal_MA = df['MA5'] > df['MA20']
df['Signal_MA'] = Signal_MA
print(df[['日期','收盘','MA5','MA20','Signal_MA']].tail())
df.head(20)
```

```python
df.dropna(inplace=True)
df.head()
```

```python
delta = df['收盘'].diff()
gain = delta.where(delta>0,0)
loss = -delta.where(delta<0,0)
avg_gain = gain.rolling(window=14).mean()
avg_loss = loss.rolling(window=14).mean()
rs = avg_gain/avg_loss
RSI = 100-(100/(1+rs))
df['Signal_RSI_Weak'] = RSI < 30
df['Signal_RSI_Hot'] = RSI > 70
df.dropna(inplace=True)
```

```python
df['DIF'] = df['收盘'].ewm(span=12,adjust=False).mean() - df['收盘'].ewm(span=26,adjust=False).mean()
df['DEA'] = df['DIF'].ewm(span=9,adjust=False).mean()
df['Signal_MACD_Bull']= df['DIF'] >df['DEA']
columns_to_drop = ['DIF', 'DEA']
df.drop(columns_to_drop, axis=1, inplace=True)
df.dropna(inplace=True)
print(df.tail())
```

```python
df['Target'] = (df['涨跌幅']>0).shift(-1)
df.dropna(inplace=True)
df.tail(3)
```

```python
from sklearn.model_selection import train_test_split
X = df[['Signal_MA', 'Signal_RSI_Weak', 'Signal_RSI_Hot', 'Signal_MACD_Bull']]
y = df['Target'].astype(int)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,shuffle=False)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100,random_state=42)
model.fit(X_train,y_train)
accuracy = model.score(X_test,y_test)
print(f"准确率：{accuracy * 100:.2f}%")
```
